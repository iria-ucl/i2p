```python
# Replace the column names with something more informative

happiness.columns = ['Country', 'Region', 'Hemisphere', 'HappinessScore', 'HDI', 'GDP_PerCapita', 'Beer_PerCapita', 'Spirit_PerCapita', 'Wine_PerCapita']
happiness.head(5)
```

```python
# find out the type of all variables within the dataset happiness, using a single line of code


happiness.dtypes
```

```python
happiness['HappinessScore'] = happiness['HappinessScore'].apply(lambda x: x.replace(',', '.'))
happiness['GDP_PerCapita'] = happiness['GDP_PerCapita'].apply(lambda x: x.replace(',', '.'))
```

```python
happiness.dtypes
```

```python
# change the variable type of the columns 'HappinessScore' and 'GDP_PerCapita' to numeric

happiness['HappinessScore'] = happiness['HappinessScore'].astype(float)
happiness['GDP_PerCapita'] = happiness['GDP_PerCapita'].astype(float)

happiness.dtypes
```

```python
# Correct the typos and multiple categories of the variable 'Hemisphere' so it will only contain three possible values north, south or both

happiness.Hemisphere.unique()
```

```python
happiness.Hemisphere = happiness.Hemisphere.replace(['North', 'N', 'noth', 'nort', 'n'],'north')
happiness.Hemisphere = happiness.Hemisphere.replace(['s','S', 'South', 'sout', 'southh'], 'south')
happiness.Hemisphere.unique()
```

```python
happiness.head()
```

```python
# Count how many observations are in the happiness dataset using the function .count(); save that number in a variable called totalCounts

totalcount = happiness['Country'].count()
print(totalcount)
```

```python
# Calculate the sum of all the Happiness score for all countries, save that value in a variable called summation

sumHappiness = happiness['HappinessScore'].sum()
print(sumHappiness)
```

```python
# Calculate the mean with the two numbers above

x = sumHappiness / totalcount
x
```

```python
# Import the library seaborn as sns


import seaborn as sns

sns.boxplot(data=happiness[['Beer_PerCapita','Spirit_PerCapita', 'Wine_PerCapita']])
plt.title('Country Alcohol Consumption') # add a title
plt.ylabel('Litres per capita (l)') # add a label for the y-axis
plt.xticks(ticks = (0,1,2),labels =['Beer', 'Spirit', 'Wine'])  # add labels for the x-axis
plt.show()
```

```python
# Create a vertical violin plot of GDP_PerCapita, separated by Region, change the title and the labels accordingly.

sns.violinplot(x= 'Hemisphere', y = 'GDP_PerCapita',
               order = ('north', 'both', 'south'), data=happiness)
plt.title('Gross domestic product per capita')
plt.ylabel('GDP per capita')
plt.xlabel('Region')
plt.xticks(ticks = (0,1,2),labels =('North', 'Both', 'South'))
plt.show()
```
